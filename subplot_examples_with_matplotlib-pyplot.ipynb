```python
from matplotlib import pyplot as PLT
```

#### Let's declare a function that plots a 2x1 figure (2 rows and 1 column) with 2 subplots

```python
def two_by_one_plot():
    fig = PLT.figure()

    ax1 = fig.add_subplot(211) # (211) = (2,1,1) => 2 rows, 1 column, 1st plot
    ax1.plot([(1, 2), (3, 4)], [(4, 3), (2, 3)])

    ax2 = fig.add_subplot(212) # (212) = (2,1,2) => 2 rows, 1 column, 2nd plot
    ax2.plot([(7, 2), (5, 3)], [(1, 6), (9, 5)])

    # To show it outside jupyter notebook
    PLT.show()
```

```python
two_by_one_plot()
```

#### Now a 2x2 plot function with 4 subplots

```python
def two_by_two():
    fig = PLT.figure()

    ax1 = fig.add_subplot(221) # upper left plot
    ax1.plot([(1, 2), (3, 4)], [(4, 3), (2, 3)])

    ax2 = fig.add_subplot(222) # upper right plot
    ax2.plot([(7, 2), (5, 3)], [(1, 6), (9, 5)])

    ax3 = fig.add_subplot(223) # lower left plot
    ax3.plot([(7, 4), (5, 2)], [(4, 1), (5, 6)])

    ax4 = fig.add_subplot(224) # lower right plot
    ax4.plot([(4, 1), (3, 7)], [(4, 4), (6, 3)])

    # To show it outside jupyter notebook
    PLT.show()
```

```python
two_by_two()
```

#### Now let's randomize the values

```python
import random as rand
```

Lower and upper bounds to simulate our data

```python
lower_bound = 0
upper_bound = 9
```

#### Helper functions to simulate some data

Random value between our `lower_bound` and `upper_bound`

```python
def random_int_value_between_upper_and_lower_bound():
    return rand.randint(lower_bound, upper_bound)
```

Tuple with 2 random values

```python
def random_tuple():
    return (random_int_value_between_upper_and_lower_bound(), random_int_value_between_upper_and_lower_bound())
```

List of random tuples

```python
def random_list_of_tuples():
    return [random_tuple(), random_tuple()]
```

##### Same functions as before, but now using random values

```python
def two_by_one_plot():
    fig = PLT.figure()

    ax1 = fig.add_subplot(211) # (211) = (2,1,1) => 2 rows, 1 column, 1st plot
    ax1.plot(random_list_of_tuples(), random_list_of_tuples()) # Using our functions to simulate the data

    ax2 = fig.add_subplot(2,1,2) # (212) = (2,1,2) => 2 rows, 1 column, 2nd plot
    ax2.plot(random_list_of_tuples(), random_list_of_tuples()) # Using our functions to simulate the data

    # To show it outside jupyter notebook
    PLT.show()
```

```python
two_by_one_plot() # Now everytime we execute this line it will (most likely) generate a different result
```

### Now we can refactor the code to make iterable

#### Let's define just one function that does all the magic

We can declare 3 parameters to help us out:
1. `number_of_rows` - Number of rows in the figure
2. `number_of_columns` - Number of columns in the figure
3. `number_of_plots` - Number of plots to show. Has to be > 0

```python
def n_by_n_plot(number_of_rows, number_of_columns, number_of_plots):
    fig = PLT.figure()

    # Let's iterate over the number of plots we want
    for current_plot in range(1, number_of_plots+1): # Number of plots can't be 0
        # Remember that calling fig.add_subplot(211) is the same as fig.add_subplot(2,1,1) for example
        axis = fig.add_subplot(number_of_rows, number_of_columns, current_plot)
        axis.plot(random_list_of_tuples(), random_list_of_tuples())

    # Show it only once, outside the for loop
    # To show it outside jupyter notebook
    PLT.show()
```

Now let's compare the previous implementations with the new one

#### Okay, that last one was a mess. We can always adjust the size of the figure

Let's add a new parameter called `figure_size` to determine the Figure size, which will be a tuple with 2 values: `width` and `height`.

```python
def n_by_n_plot(number_of_rows, number_of_columns, number_of_plots, figure_size):
    fig = PLT.figure(figsize=figure_size) # figsize must be a tuple, documentation: https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html

    # Let's iterate over the number of plots we want
    for current_plot in range(1, number_of_plots+1): # Number of plots can't be 0
        # Remember that calling fig.add_subplot(211) is the same as fig.add_subplot(2,1,1) for example
        axis = fig.add_subplot(number_of_rows, number_of_columns, current_plot)
        axis.plot(random_list_of_tuples(), random_list_of_tuples())

    # Show it only once, outside the for loop
    # To show it outside jupyter notebook
    PLT.show()
```

```python
n_by_n_plot(10, 1, 10, (10, 20))
```

```python
n_by_n_plot(5, 2, 10, (20, 10))
```
